**Basic Plotting using Bokeh**

```python
#import necessary libraries

from bokeh.plotting import figure
from bokeh.io import output_file, show, output_notebook, push_notebook
from bokeh.models import ColumnDataSource, HoverTool, CategoricalColorMapper, Tabs, Panel
from bokeh.layouts import row, column, gridplot
#from bokeh.models.widgets import Tabs, Panel
output_notebook()
```

```python
f = figure()
f.line([10,20,30,40,50],[60,70,80,40,50],line_width=12)

show(f)
```

**Plotting using Glyph**



```python
# Circle Glyph

p = figure(x_axis_label = "x axis",
           y_axis_label="y axis",
           tools ="pan, box_zoom")
p.circle(x=[50,40,30],
         y=[10,20,30],
         size=25,
         color="red",
         alpha=0.7,
         line_color="yellow")

show(p)
```

```python
# Square Glyph

t = figure(width=300,
           height=500)

t.square([10,12,13,14, 15],
         [16,17, 12, 14, 15],
         size=20,
         color="blue",
         alpha=0.7)

show(t)
```

```python
# Multiple Lines

y = figure(width=400,
           height=500)
y.multi_line([[11,13,12], [31,41, 16]],
             [[11,12,41], [14,71, 18] ],
             color=["red","blue"],
             alpha=[0.8,0.3],
             line_width=5)
show(y)
```

```python
# Wedge Glyph

z= figure(width=400,
          height=500)

z.wedge(x=[1,2,3],
        y=[1,2,3],
        radius=0.2,
        start_angle=0.4,
        end_angle=5.2,
        color="black",
        alpha=0.7,
        direction="clock")

show(z)
```

```python
# Annular Wedge Glyph

q =figure(width=500,height=600)

q.annular_wedge(x=[10,20,30],
                y=[10,20,30],
                inner_radius=0.5, outer_radius=0.75,
                start_angle=0.5,end_angle=5.2,
                color="green", alpha=0.7)

show(q)
```

```python
# Combine Circle and Line on the same plot

x=[10,20,30,40]
y=[60,75,80,90]

u = figure(width=400,height=500)

u.line(x,y,line_width=2)
u.circle(x,y, fill_color="black",size=9)

show(u)
```

```python
# Bar Plot

g1= figure(title= " Bar Plot")

x = [1,2,3,4,5]
y = [1,2,3,4,5]

g1.hbar(x,right=y,height=0.6)
show(g1)
```

```python
# Vertical Bar Plot

g2= figure(title="Vertical Bar Plot")

x=[10,20,30]
y=[40,50,60]

g2.vbar(x,top=y,width=0.6)

show(g2)
```

```python
# Scatter Plot

import random

g3=figure(title="Scatter Plot")

x=[i for i in range(126)]
y=[random.random() + 2 for i in range(126)]

g3.scatter(x,y)

show(g3)
```

```python
# Area Plot

import numpy as np

x=[10,20,30,40,50]
y1=[2,4,7,8,9]
y2=[1,2,3,4,5]

h= figure(width=400,
          height=400)

h.varea(x=x,
        y1=y1,
        y2=y2,
        fill_color="blue")
show(h)
```

```python
# Patch Plot

w= figure(width=400,
          height=400)

w.patch(x=[1,4,6,9],
        y=[2,3,7,8],
        color="navy")

show(w)
```
